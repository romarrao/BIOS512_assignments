```r
library('tidyverse')
```

```r
wind_data = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',
                     col_types = cols()) #load raw data

wind_data |> head()
```

Each row in the table represents a single turbine.

Make a bar chart üìä **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_data = wind_data |> 
    mutate(province = province_territory |> 
           fct_infreq() |>
           fct_rev()) 
```

```r
wind_data |> pull(province) |> levels()
```

```r

options(repr.plot.res = 200)
p = ggplot(wind_data, aes(x=province)) + 
    geom_bar(fill='purple') +
    theme(axis.text.x = element_text(angle = 45, hjust=1)) +
    labs(x='Province Territory', y='Turbine Count', title='Frequency of Turbine Count per Province Territory')
    
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
manufacturer_col = wind_data |> 
    group_by(manufacturer) |>
    pull(manufacturer) |>
    fct_infreq() |> 
    fct_rev() |> 
    fct_lump(n=7) # increased from 5 to have more manufacturers on x-axis

manufacturer_col |> head()
```

```r
options(repr.plot.res = 200)
p = ggplot(wind_data, aes(x=manufacturer_col)) +
    geom_bar(fill='steelblue') +
    theme(axis.text.x = element_text(angle = 45, hjust=1)) +
    labs(x='Manufacturer', y='Turbine Count', title='Frequency of Turbine Count per Manufacturer') 

p
```
